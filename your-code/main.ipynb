#### 2. Import users table:

#### 4. Import posts table:

```python
posts_table = pd.read_csv ('posts_table.csv')
posts_table.head()
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts_table.rename(columns={'Id':'postId', 'OwnerUserId': 'userId'}, inplace = True)
posts_table.head()
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
filtered_users = users_table[['userId','Reputation','Views','UpVotes','DownVotes']] 
filtered_posts = posts_table[['postId','Score','userId','ViewCount','CommentCount']] 
filtered_posts.head(20)
```
